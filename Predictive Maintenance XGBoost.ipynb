```python
from sklearn.model_selection import KFold, train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix, mean_absolute_error,accuracy_score


x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.7, random_state=123)
eval_set = [(x_test,y_test)]
```

```python
modelXGB= XGBClassifier(objective ='reg:linear', 
                             max_depth = 3,
                             silent = 1,
                             learning_rate = 0.3,
                             n_estimators = 200)
```

```python
model=modelXGB.fit(x_train, y_train, early_stopping_rounds=10,eval_metric= 'logloss', eval_set=eval_set)
```

```python
y_pred = model.predict(x_test)
print("Classification Report: \n", classification_report(y_test, y_pred))
print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred))
accuracy = accuracy_score(y_test, y_pred)
print("\nAccuracy",accuracy)
```
